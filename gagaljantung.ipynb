## Prediksi Gagal Jantung (Support Vector Machine)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import pickle
```

```python
dataset=pd.read_csv('heart_failure_clinical_records_dataset.csv')
df=dataset
```

```python
df.head()
```

```python
df.info()
```

```python
cols= ["#CD5C5C","#FF0000"]
ax = sns.countplot(x= df["label"], palette= cols)
ax.bar_label(ax.containers[0])
```

```python
cmap = sns.diverging_palette(2, 165, s=80, l=55, n=9)
corrmat = df.corr()
plt.subplots(figsize=(20,20))
sns.heatmap(corrmat,cmap= cmap,annot=True, square=True)
```

```python
df.shape
```

```python
X=df.drop(columns='label',axis=1)
y=df['label']
```

```python
print(X)
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=svm.SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
X_train_predict=model.predict(X_train)
train_data_accuracy=accuracy_score(X_train_predict,y_train)
print('Akurasi Data Latih : ',train_data_accuracy)
```

```python
X_test_predict=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_predict,y_test)
print('Akurasi Data Test : ',test_data_accuracy)
```

### Membuat Model Prediksi

```python
input_data=[[75,0,582,0,20,1,265000,1.9,130,1,0,4]]

prediction=model.predict(input_data)
print(prediction)

if(prediction==0):
  print('Pasien Relatif Tidak Beresiko')
else:
  print('Pasien Beresiko Gagal Jantung')
```

```python
filename='gagaljantung-model.sav'
pickle.dump(model,open(filename,'wb'))
```
